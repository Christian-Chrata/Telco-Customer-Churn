# Import Libraries

```python
import pandas as pd
import pickle
```

# Read

```python
with open("tuned_logistic_regression.pkl", "rb") as file:
    loaded_model = pickle.load(file)

new_data = pd.read_csv("dataset/df_clean.csv")
```

# Run Model

```python
predictions = loaded_model.predict(new_data)
probabilities = loaded_model.predict_proba(new_data)[:, 1]

new_data["Churn_Predicted"] = predictions
new_data["Churn_Predicted"] = new_data["Churn_Predicted"].map({0: "No", 1: "Yes"})
new_data["Churn_Probability"] = probabilities
```

# Filter Churn Predicted Customer

```python
churn_list = new_data[new_data["Churn_Predicted"] == "Yes"]
churn_list.sample(5)
```

# Export to CSV

```python
churn_list[["customerID", "Churn_Probability", "Churn_Predicted"]].to_csv("Dataset/list_predicted_customer.csv", index=False)
```

```python
detail = churn_list.drop(columns=["Churn_Probability", "Churn_Predicted", "Churn"])
detail = detail.rename(columns={"customerID": "Customer ID"})


detail.to_csv("Dataset/detail_predicted_customer.csv", index=False)
```
